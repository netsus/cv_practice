```python
import cv2
from matplotlib import pyplot as plt
```

```python
def imshow(img, figsize=(10,10)):
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.figure(figsize=figsize)
    plt.imshow(img)
    plt.show()
```

```python
img=cv2.imread('Lenna.png', cv2.IMREAD_COLOR)
line_img = cv2.line(
    img,
    pt1=(0,0),
    pt2=(img.shape[0], img.shape[1]),
    color=(255,0,0),
    thickness=3,
    lineType=cv2.LINE_AA
)
imshow(img)
```

```python
img=cv2.imread('Lenna.png', cv2.IMREAD_COLOR)
circle_img = cv2.circle(
    img,
    radius=140,
    center=(img.shape[1]//2, img.shape[0]//2), # col, row   - shape(,)  x,  y
    color=(255,0,0),
    thickness=3,
#     lineType=cv2.LINE_AA
)
imshow(circle_img)
```

```python
img=cv2.imread('Lenna.png', cv2.IMREAD_COLOR)
circle_img = cv2.rectangle(
    img,
    pt1=(100,50), #  (x,y)  0,0
    pt2=(300,400), # 
    color=(0,255,0),
    thickness=3,
#     lineType=cv2.LINE_AA
)
imshow(circle_img)
```

```python
img=cv2.imread('Lenna.png', cv2.IMREAD_COLOR)
circle_img = cv2.putText(
    img,
    text="CarpeDiem",
    org = (100,100), # origin    
    fontScale = 1.9,
    fontFace = cv2.FONT_HERSHEY_SIMPLEX,
    color=(195,255,150),
    thickness=4,
    lineType=cv2.LINE_AA
)
imshow(circle_img)
```
